```python
# PROJETO 2 - JOGO DA VELHA - PROF.: Brian| Luis | Larissa

# Range 10 utilizado para poder desconsiderar o indice 0
tabuleiro = [' ' for x in range(10)]

# Faz a jogada escolhida
def insereLetra(letra, posicao):
    tabuleiro[posicao] = letra


# Verifica a disponibilidade
def espacoDisponivel(posicao):
    return tabuleiro[posicao] == ' '


# Layout tabuleiro 3 x 3 em posicoes determinadas de 1 2 3
#                                                    4 5 6
#                                                    7 8 9
def mostraTabuleiro(tabuleiro):
    print('      |     |')
    print(' ' + tabuleiro[1] + '    | ' + tabuleiro[2] + '   | ' + tabuleiro[3])
    print('      |     |')
    print('-------------------')
    print('      |     |')
    print(' ' + tabuleiro[4] + '    | ' + tabuleiro[5] + '   | ' + tabuleiro[6])
    print('      |     |')
    print('-------------------')
    print('      |     |')
    print(' ' + tabuleiro[7] + '    | ' + tabuleiro[8] + '   | ' + tabuleiro[9])
    print('      |     |\n')


# Condicoes de vitoria considerando a ordem que as posicoes foram inseridas
def ehVencedor(position, opcao):
    return (position[1] == opcao and position[2] == opcao and position[3] == opcao)\
    or (position[1] == opcao and position[3] == opcao and position[2] == opcao)\
    or (position[2] == opcao and position[3] == opcao and position[1] == opcao)\
    or (position[2] == opcao and position[1] == opcao and position[3] == opcao)\
    or (position[3] == opcao and position[2] == opcao and position[1] == opcao)\
    or (position[3] == opcao and position[1] == opcao and position[2] == opcao)\
    or (position[4] == opcao and position[5] == opcao and position[6] == opcao)\
    or (position[4] == opcao and position[6] == opcao and position[5] == opcao)\
    or (position[5] == opcao and position[4] == opcao and position[6] == opcao)\
    or (position[5] == opcao and position[6] == opcao and position[4] == opcao)\
    or (position[6] == opcao and position[4] == opcao and position[5] == opcao)\
    or (position[6] == opcao and position[5] == opcao and position[4] == opcao)\
    or (position[7] == opcao and position[8] == opcao and position[9] == opcao)\
    or (position[7] == opcao and position[9] == opcao and position[8] == opcao)\
    or (position[8] == opcao and position[9] == opcao and position[7] == opcao)\
    or (position[8] == opcao and position[7] == opcao and position[9] == opcao)\
    or (position[9] == opcao and position[8] == opcao and position[7] == opcao)\
    or (position[9] == opcao and position[7] == opcao and position[8] == opcao)\
    or (position[1] == opcao and position[4] == opcao and position[7] == opcao)\
    or (position[1] == opcao and position[7] == opcao and position[4] == opcao)\
    or (position[2] == opcao and position[5] == opcao and position[8] == opcao)\
    or (position[2] == opcao and position[8] == opcao and position[5] == opcao)\
    or (position[5] == opcao and position[2] == opcao and position[8] == opcao)\
    or (position[5] == opcao and position[8] == opcao and position[2] == opcao)\
    or (position[8] == opcao and position[5] == opcao and position[2] == opcao)\
    or (position[8] == opcao and position[2] == opcao and position[5] == opcao)\
    or (position[3] == opcao and position[6] == opcao and position[9] == opcao)\
    or (position[3] == opcao and position[9] == opcao and position[6] == opcao)\
    or (position[6] == opcao and position[3] == opcao and position[9] == opcao)\
    or (position[6] == opcao and position[9] == opcao and position[3] == opcao)\
    or (position[9] == opcao and position[3] == opcao and position[6] == opcao)\
    or (position[9] == opcao and position[6] == opcao and position[3] == opcao)\
    or (position[1] == opcao and position[5] == opcao and position[9] == opcao)\
    or (position[1] == opcao and position[9] == opcao and position[5] == opcao)\
    or (position[5] == opcao and position[1] == opcao and position[9] == opcao)\
    or (position[5] == opcao and position[9] == opcao and position[1] == opcao)\
    or (position[9] == opcao and position[5] == opcao and position[1] == opcao)\
    or (position[9] == opcao and position[1] == opcao and position[5] == opcao)\
    or (position[3] == opcao and position[5] == opcao and position[7] == opcao)\
    or (position[3] == opcao and position[7] == opcao and position[5] == opcao)\
    or (position[5] == opcao and position[3] == opcao and position[7] == opcao)\
    or (position[5] == opcao and position[7] == opcao and position[3] == opcao)\
    or (position[7] == opcao and position[3] == opcao and position[5] == opcao)\
    or (position[7] == opcao and position[5] == opcao and position[3] == opcao)


# Movimento do usuario com checagem de espaco disponivel e entrada correta
def movimentoJogador():
    run = True
    while run:
        move = input('Faça seu movimento com um \'O\' entre as posições (1-9): ')
        try:
            move = int(move)
            if move > 0 and move < 10:
                if espacoDisponivel(move):
                    run = False
                    insereLetra('O', move)
                else:
                    print('Movimento já realizado')
            else:
                print('Digite um número de 1 a 9 para uma posicao vazia!')
        except:
            print('Digite um número de 1 a 9 para uma posicao vazia: ')
# Movimento da maquina conferindo a possibilidade e eficiencia do movimento de acordo com os criterios abaixo
# estabelecidos: vencer, não perder, quinas, centro e meio das laterais.
# Esses criterios foram estabelecidos apos analisar as 48 formas de vencer o jogo.
def movimentoMaquina():
    movimentosEficientes = [x for x, letra in enumerate(tabuleiro) if letra == ' ' and x != 0]
    move = 0

# Busca a vitoria ou impedir a derrota
    for let in ['X', 'O']:
        for i in movimentosEficientes:
            tabuleiroCopia = tabuleiro[:] # copia toda a lista
            tabuleiroCopia[i] = let
            if ehVencedor(tabuleiroCopia, let):
                move = i
                return move
    
    quinasAbertos = []
    for i in movimentosEficientes:
        if i in [1,3,7,9]:
            quinasAbertos.append(i)
            
    if len(quinasAbertos) > 0:
        move = selectRandom(quinasAbertos)
        return move

    if 5 in movimentosEficientes:
        move = 5
        return move

    lateraisAbertas = []
    for i in movimentosEficientes:
        if i in [2,4,6,8]:
            lateraisAbertas.append(i)
            
    if len(lateraisAbertas) > 0:
        move = selectRandom(lateraisAbertas)
        
    return move


# Sera usada para "randomizar" o movimento mas com critérios
def selectRandom(li):
    import random
    ln = len(li)
    r = random.randrange(0,ln)
    return li[r]
    


def tabuleiroCheio(tabuleiro):
    if tabuleiro.count(' ') > 1:
        return False
    else:
        return True


# Funcao responsavel por ordenar todo o codigo
def estruturaDoJogo():
    print('JOGO DA VELHA')
    mostraTabuleiro(tabuleiro)

    while not(tabuleiroCheio(tabuleiro)):
        if not(ehVencedor(tabuleiro, 'O')):
            move = movimentoMaquina()
            if move == 0:
                print('Empate!')
            else:
                insereLetra('X', move)
                print('O computador marcou um \'X\' na posição', move , ':')
                mostraTabuleiro(tabuleiro)
        else:
            print('O\' Parabéns...você venceu!')
            break
             
        if not(ehVencedor(tabuleiro, 'X')):
            movimentoJogador()
            mostraTabuleiro(tabuleiro)
        else:
            print('Você perdeu!')
            break



    if tabuleiroCheio(tabuleiro):
        print('Fim do Jogo')

while True:
    jogo = input('Vamos jogar o Jogo da Velha? (S/N)')
    if jogo.upper() == 'S' or jogo.upper() == 's':
        tabuleiro = [' ' for x in range(10)]
        print('-----------------------------------')
        estruturaDoJogo()
    else:
        break
```
