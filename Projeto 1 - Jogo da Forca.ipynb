```python
from random import choice

def titulo(tit):
    print('-' * 30)
    print(tit)
    print('-' *30)
    
    
titulo('         Jogo da Forca  ')

def modoSolo():
        
    frutasEverduras = [
        'abacate', 'abacaxi', 'acerola', 'ameixa', 'amora', 'banana', 'caju',
        'coco', 'damasco', 'figo', 'framboesa', 'goiaba', 'graviola', 'imbu', 
        'alface', 'brocolis', 'tomate', 'cenoura', 'batata', 'beterraba', 'quiabo',
        'vagem', 'milho', 'ervilha'
    ]

    # Colocando os underlines = número de letras da palavra
    palavra = choice(frutasEverduras)
    palavraCorreta = list(palavra)
    segredo = []
    for letra in palavra:
        segredo.append('_')

    tentativas = 0
    maxTentativas = 4

    # Usar while para manter o usuário interagindo até perder ou ganhar
    fimDoJogo = False
    while not fimDoJogo:

        # Aqui é um quadro com a situação completa e atualizada do jogo
        print(f"Você ainda tem {maxTentativas - tentativas} tentativas")
        # Mostrar o número de letras da palavra
        print(f"A palavra é: {' '.join(segredo)}")
        
        print("     ------")
        print("     |    |")
        print("     |    " + ("O" if tentativas > 0 else ""))
        print("     |    " + ("/\\" if tentativas > 1 else ""))
        print("     |    " + ("|" if tentativas > 2 else ""))
        print("     |    " + ("/\\" if tentativas > 3 else ""))
        print(" --------")


        letraEscolhida = input("Escolha uma letra: ")

        print('\n\n\n')

        if letraEscolhida in palavraCorreta:
            print(f"Escolheu certo! {letraEscolhida} está na palavra.")
            for i in range(len(palavraCorreta)):
                letra = palavra[i]
                if letra == letraEscolhida:
                    segredo[i] = palavraCorreta[i]
                    palavraCorreta[i] = '_'
        else:
            print(f"Escolheu mal! {letraEscolhida} não está na palavra.")
            tentativas += 1
                    
        if all('_' == letra for letra in palavraCorreta):
            print("Parabéns, você venceu!")
            fimDoJogo = True
            

        if tentativas >= maxTentativas:
            print("Fim do Jogo!")
            fimDoJogo = True
    

def modoVersus():
        
    # Colocando os underlines = número de letras da palavra
    palavra = str(input('Escolha uma palavra qualquer (não digite acento caso tenha)'))
    palavraCorreta = list(palavra)
    segredo = []
    for letra in palavra:
        segredo.append('_')
    
    tentativas = 0
    maxTentativas = 4

    # Usar while para manter o usuário interagindo até perder ou ganhar
    fimDoJogo = False
    while not fimDoJogo:

        # Aqui é um quadro com a situação completa e atualizada do jogo
        print(f"Você ainda tem {maxTentativas - tentativas} tentativas")
        # 
        print(f"A palavra é: {' '.join(segredo)}")
        
        print("     ------")
        print("     |    |")
        print("     |    " + ("O" if tentativas > 0 else ""))
        print("     |    " + ("/\\" if tentativas > 1 else ""))
        print("     |    " + ("|" if tentativas > 2 else ""))
        print("     |    " + ("/\\" if tentativas > 3 else ""))
        print(" --------")


        letraEscolhida = input("Escolha uma letra: ")

        print('\n\n\n')

        if letraEscolhida in palavraCorreta:
            print(f"Escolheu certo! {letraEscolhida} está na palavra.")
            for i in range(len(palavraCorreta)):
                letra = palavra[i]
                if letra == letraEscolhida:
                    segredo[i] = palavraCorreta[i]
                    palavraCorreta[i] = '_'
    
        else:
            print(f"Escolheu mal! {letraEscolhida} não está na palavra.")
            tentativas += 1
            fimDoJogo = False

        if all('_' == letra for letra in palavraCorreta):
            print("Parabéns, você venceu!")
            fimDoJogo = True


        if tentativas == maxTentativas:
            print(f" A palavra era {palavra}. \nFim do Jogo!")
            fimDoJogo = True
        


menu = 'ESCOLHA O MODO DO JOGO:\n'
menu += '1 - Modo Solo - Será usada uma lista de frutas e verduras;\n'
menu += '2 - Modo Versus - Seu oponente escolhe uma palavra para você acertar;\n'
menu += '3 - Sair do jogo;\n'


opcao = int(input(menu))

if opcao == 1:
    modoSolo()


elif opcao == 2:
    modoVersus()
    
else: 
    print('Obrigado!')
```
